# Extrai os metadados necessários das imagens.

```
from PIL import Image
from PIL.ExifTags import TAGS
import os
import pprint as pp
```

```
cwd = os.getcwd()
```

##### Arquivo onde serão escritas as informações

```
f = open("metadata.txt", "w")
imgs = os.listdir("a")
imgs
```

##### Escrita das informações.

```
list = []
for i in imgs:
    image = Image.open(cwd + "/a/"+ i)
    width, height = image.size
    list.append([width, height])
    pp.pprint(f"{i}: [width: {width}, height: {height}]")
```

```
f.close()
```
